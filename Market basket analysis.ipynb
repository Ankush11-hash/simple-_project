# Market basket analysis of data in list form

Q) Find the items with at least 60% support.

Which three itemset is a sure shot indicator that they will maximise profits if kept together

Dataset (Data is in the form of Python list)

```python

import numpy as np
import pandas as pd
from apyori import apriori
dataset = [["Shampoo", 'Conditioner',' Soap', 'Shower Gel', 'Facewash',
'Shaving Foam'],
['Body Lotion','Soap','Aftershave', 'Facewash', 'Shower Gel', 'Shaving Foam'],
['Shampoo', 'Soap', 'Shower Gel', 'Facewash'],
['Conditioner', 'Shaving Foam', 'Soap', 'Aftershave', 'Shower Gel'],
['Soap', 'Shampoo', 'Aftershave', 'Shower Gel','Facewash'],
['Shower Gel', 'Soap', 'Facewash', 'Body Lotion'],
['Body Lotion', 'Shampoo', 'Facewash'],
['Body Lotion', 'Aftershave', 'Soap', 'Shower Gel', 'Facewash']]
dataset
```

```python
d = list(apriori(dataset))
df = pd.DataFrame(d)
df
```

```python
df_sort = df.loc[df['support']>=.6]
df_sort = df_sort.reset_index(drop=True)
df_sort
# index 6 shows support higher than 60 percent 
```

# Above data frame Shows set of single , 2 items and 3 items which have support greater than 60 percent. 

**Soap, Facewash, Shower Gel has support of 62.5%** and if placed together it could boost profit of the retial store/outlet
