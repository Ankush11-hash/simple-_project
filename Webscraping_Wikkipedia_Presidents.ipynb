**Scrape down the list of Presidents from the link mentioned below** :https://en.wikipedia.org/wiki/List_of_Presidents_of_the_United_States

```
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
```

```
link = 'https://en.wikipedia.org/wiki/List_of_Presidents_of_the_United_States'
text = requests.get(link)
print(text.status_code)# 200 means we can access the page and 404 error ocure when content is not accessbile
soup  = bs(text.text,'lxml')
```

```
table = soup.find('table',{'class':"wikitable"})
print(table.prettify())
```

```
a= table.findAll('b')
President = []
for i in range(len(a)):
  President.append(a[i].text)

df = pd.DataFrame({'President':President})
df
```
